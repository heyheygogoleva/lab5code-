## Используя паттерн Builder, реализовать процесс создания объекта типа "Пицца" разных видов

```python
class Pizza:
    def __init__(self):
        self.without_meat = False
        self.with_meat = False
        self.mushrooms = False
        self.cheese = []
        self.vegetables = []
        self.meat = []
        self.special = []

    def __str__(self):
        string = ""
        if self.without_meat:
            string += "without any meat "
        if self.with_meat:
            string += "with maaany meat "
        if self.mushrooms:
            string += "with mushrooms "
        else:
            string += "pizza\n"

        if self.cheese:
            string += "Added cheese:\n"
        for product in self.cheese:
            string += "- " + str(product) + "\n"    
            
        if self.vegetables:
            string += "Added vegetables:\n"
        for product in self.vegetables:
            string += "- " + str(product) + "\n"

        if self.meat:
            string += "Added meat:\n"
        for product in self.meat:
            string += "- " + str(product) + "\n"
            
        if self.special:
            string += "Added specialties:\n"
        for product in self.special:
            string += "- " + str(product) + "\n"

        return string

class NonMilkCheese:
    def __str__(self):
        return "lactose-free cheese"

class MozzarellaCheese:
    def __str__(self):
        return "mozzarella"

class MoldCheese:
    def __str__(self):
        return "cheese with mold"

class CheddarCheese:
    def __str__(self):
        return "cheddar cheese"

class ChickenMeat:
    def __str__(self):
        return "chicken"

class BeefMeat:
    def __str__(self):
        return "beef"

class DuckMeat:
    def __str__(self):
        return "duckling" # утенок

class BaconMeat:
    def __str__(self):
        return "bacon meat"

class BroccoliVegetable:
    def __str__(self):
        return "broccoli"

class SweetPepperVegetable:
    def __str__(self):
        return "sweet peppers"

class TomatosVegetable:
    def __str__(self):
        return "tomatos"


class OnionVegetable:
    def __str__(self):
        return "onion"

class CornVegetable:
    def __str__(self):
        return "sweet corn"

class ChilliVegetable:
    def __str__(self):
        return "chilli"

class PinappleVegetable:
    def __str__(self):
        return "pinapples"
    
class EggplantVegetable:
    def __str__(self):
        return "eggplants"
    
    
class MushroomsVegetable:
    def __str__(self):
        return "mushrooms"
```

```python
# Абстрактные базовые классы (ABC) определяют, какие производные классы реализуют конкретные методы из базового класса.

from abc import ABC, abstractmethod

class PizzaMaker(ABC):

    @abstractmethod
    def reset(self):
        pass

    @abstractmethod
    def add_cheese(self):
        pass
    
    @abstractmethod
    def add_vegetables(self):
        pass

    @abstractmethod
    def add_meat(self):
        pass
    
    @abstractmethod
    def add_special(self):
        pass
    
```

```python
class VegetarianMaker(PizzaMaker):
    def __init__(self):
        self.product = Pizza()

    def reset(self):
        self.product = Pizza()

    def get_product(self):
        return self.product

    def add_vegetables(self):
        self.product.without_meat = True
        self.product.vegetables.append(BroccoliVegetable())
        self.product.vegetables.append(SweetPepperVegetable())
        self.product.vegetables.append(OnionVegetable())
        self.product.vegetables.append(TomatosVegetable())

    def add_cheese(self):
        self.product.cheese.append(MozzarellaCheese())
    
    def add_meat(self):
        pass
    
    def add_special(self):
        pass
    
    

class MeatMaker(PizzaMaker):
    def __init__(self):
        self.product = Pizza()

    def reset(self):
        self.product = Pizza()

    def get_product(self):
        return self.product

    def add_vegetables(self):
        self.product.vegetables.append(TomatosVegetable())

    def add_cheese(self):
        self.product.cheese.append(MozzarellaCheese())
    
    def add_meat(self):
        self.product.with_meat = True
        self.product.meat.append(ChickenMeat())
        self.product.meat.append(BeefMeat())
        self.product.meat.append(DuckMeat())
        self.product.meat.append(BaconMeat())
        
    def add_special(self):
        pass
    
class ClassicalMaker(PizzaMaker):
    def __init__(self):
        self.product = Pizza()

    def reset(self):
        self.product = Pizza()

    def get_product(self):
        return self.product

    def add_vegetables(self):
        self.product.vegetables.append(MushroomsVegetable())

    def add_cheese(self):
        self.product.cheese.append(MozzarellaCheese())
    
    def add_meat(self):
        self.product.with_meat = True
        self.product.meat.append(BaconMeat())
        
    def add_special(self):
        pass
    
class MexicallMaker(PizzaMaker):
    def __init__(self):
        self.product = Pizza()

    def reset(self):
        self.product = Pizza()

    def get_product(self):
        return self.product

    def add_vegetables(self):
        self.product.vegetables.append(CornVegetable())

    def add_cheese(self):
        self.product.cheese.append(MozzarellaCheese)
    
    def add_meat(self):
        self.product.with_meat = True
        self.product.meat.append(BeefMeat())
        
    def add_special(self):
        self.product.special.append(ChilliVegetable())

class HawaiianMaker(PizzaMaker):
    def __init__(self):
        self.product = Pizza()

    def reset(self):
        self.product = Pizza()

    def get_product(self):
        return self.product

    def add_vegetables(self):
        pass

    def add_cheese(self):
        self.product.cheese.append(MozzarellaCheese())
    
    def add_meat(self):
        self.product.with_meat = True
        self.product.meat.append(ChickenMeat())
        
    def add_special(self):
        self.product.special.append(PinappleVegetable())
```

```python
print("- I want classical vegetarian pizza!")
print("- OK! Here it is.")
print()
cooker = VegetarianMaker()
cooker.add_vegetables()
cooker.add_cheese()
print(cooker.get_product())
```

```python
class Director:
    def make_vegetarian(self, builder):
        builder.add_vegetables()
        builder.add_cheese()
        return builder.get_product()

    def make_meat(self, builder):
        builder.add_vegetables()
        builder.add_meat()
        builder.add_cheese()
        return builder.get_product()
    
    def make_maxical(self, builder):
        builder.add_meat()
        builder.add_cheese()
        builder.add_vegetables()
        builder.add_special()
        return builder.get_product()
```
